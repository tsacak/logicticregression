```python
weekday = pd.get_dummies(df1['Weekday_Name'], drop_first=True).astype(int)
```

```python
df1 = pd.concat([df1, weekday], axis=1)
```

```python
from sklearn.model_selection import train_test_split
```

```python
X = df1.drop('Clicked on Ad', axis = 1)
y = df1['Clicked on Ad']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=101)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
logmodel = LogisticRegression()
```

```python
import warnings
warnings.filterwarnings("ignore")
```

```python
logmodel.fit(X_train, y_train)
```

```python
y_pred = logmodel.predict(X_test)
```

```python
y_test[0:5]
```

```python
y_pred[0:5]
```

```python
from sklearn.metrics import classification_report
```

```python
print(classification_report(y_test, y_pred))
```

```python
from sklearn.metrics import confusion_matrix
```

```python
confusion_matrix(y_test, y_pred)
```

```python
# Eğer Timestamp'iniz datetime formatında değilse, önce dönüştürün
df1['Timestamp'] = pd.to_datetime(df['Timestamp'])

# Gün bilgisi
df['Day_of_Week'] = df['Timestamp'].dt.day_name()

# Saat bilgisi
df['Hour'] = df['Timestamp'].dt.hour


# Gün ve saatteki tıklama sayısını bulma
clicks_per_day_hour = df.groupby(['Day_of_Week', 'Hour'])['Clicked on Ad'].sum().reset_index()
```

```python
clicks_per_day_hour
```

```python
# Gün bazında toplam tıklama sayısını hesaplayalım
clicks_per_day = df.groupby('Day_of_Week')['Clicked on Ad'].sum().reset_index()

# Sıralamak (en yüksek tıklama sayısından en düşüğe)
clicks_per_day = clicks_per_day.sort_values(by='Clicked on Ad', ascending=False)
```

```python
clicks_per_day
```

```python
# Barplot ile gün bazında tıklamaları görselleştirelim
plt.figure(figsize=(10, 6))
sns.barplot(x='Day_of_Week', y='Clicked on Ad', data=clicks_per_day, palette='viridis')
plt.title('Gün Bazında Toplam Reklama Tıklama Sayısı')
plt.xlabel('Gün')
plt.ylabel('Toplam Tıklama')
plt.show()
```

```python
# Gün ve saat bazında toplam tıklama sayısını hesaplayalım
clicks_per_day_hour = df.groupby(['Day_of_Week', 'Hour'])['Clicked on Ad'].sum().reset_index()

# Belirli bir gün için (örneğin Sunday) saat bazında tıklama sayısını filtreleyelim
sunday_clicks = clicks_per_day_hour[clicks_per_day_hour['Day_of_Week'] == 'Sunday']
```

```python
sunday_clicks
```

```python
# Pazar günü verilerini filtreleyelim
sunday_data = df[df['Day_of_Week'] == 'Sunday']

# Pazar günü saat bazında toplam tıklama sayısını hesaplayalım
clicks_per_sunday_hour = sunday_data.groupby('Hour')['Clicked on Ad'].sum().reset_index()

# Sonuçları sıralayalım (en yüksek tıklama sayısından en düşüğe)
clicks_per_sunday_hour = clicks_per_sunday_hour.sort_values(by='Clicked on Ad', ascending=False)

# Pazar günü saat dilimlerine göre tıklama sayılarını görselleştirelim
plt.figure(figsize=(10, 6))
sns.barplot(x='Hour', y='Clicked on Ad', data=clicks_per_sunday_hour, palette='viridis')
plt.title('Pazar Günü Saat Bazında Toplam Reklama Tıklama Sayısı')
plt.xlabel('Saat')
plt.ylabel('Toplam Tıklama')
plt.xticks(range(0, 24))  # Saatleri 0-23 arası göstermek için
plt.show()
```

```python
# Perşembe günü verilerini filtreleyelim
thursday_data = df[df['Day_of_Week'] == 'Thursday']

# Pazar günü saat bazında toplam tıklama sayısını hesaplayalım
clicks_per_thursday_hour = thursday_data.groupby('Hour')['Clicked on Ad'].sum().reset_index()

# Sonuçları sıralayalım (en yüksek tıklama sayısından en düşüğe)
clicks_per_thursday_hour = clicks_per_thursday_hour.sort_values(by='Clicked on Ad', ascending=False)

# Pazar günü saat dilimlerine göre tıklama sayılarını görselleştirelim
plt.figure(figsize=(10, 6))
sns.barplot(x='Hour', y='Clicked on Ad', data=clicks_per_thursday_hour, palette='viridis')
plt.title('Perşembe Günü Saat Bazında Toplam Reklama Tıklama Sayısı')
plt.xlabel('Saat')
plt.ylabel('Toplam Tıklama')
plt.xticks(range(0, 24))  # Saatleri 0-23 arası göstermek için
plt.show()
```
